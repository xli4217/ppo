## Distribution Plot

```python
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)

from rl_pipeline.visualization.viz_utils import Viz
from config import CONFIG

import os

data_dir = CONFIG['configuration'].csv_data_dir
viz = Viz()

data, layout, fig = viz.plotly_plot_distribution([data_dir], 
                                                 x_label="iteration", 
                                                 y_label="mean_return_shifted", 
                                                 std_scale=2,
                                                 legend=["legend1"],
                                                title="mean_return")
iplot(fig)
```

## Grouped Bar Plot
